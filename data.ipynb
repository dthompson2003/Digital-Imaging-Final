```python
import kagglehub
import os

# Download latest version
path = kagglehub.dataset_download("tpapp157/earth-terrain-height-and-segmentation-map-images")

print("Path to dataset files:", path)

heightMapData = []
terrainData = []


for root, dirs, files in os.walk(path):
    for file in files:
        if file.endswith("h.png"):
            full_path = os.path.join(root, file)
            heightMapData.append(full_path)
        elif file.endswith("t.png"):
            full_path = os.path.join(root, file)
            terrainData.append(full_path)

print("Number of height map files:", len(heightMapData))
print("\nFirst 10 of height map:")
for file in heightMapData[:10]:   # show first 10
    print(file)
print("-------------------------------------------------------")
print("Number of terrian image:", len(terrainData))
print("\nFirst 10 of terrian:")
for file in terrainData[:10]:   # show first 10
    print(file)
```

```python
import cv2
import numpy as np

os.makedirs("train", exist_ok=True)

for i in range(len(heightMapData)):
    h_img = cv2.imread(heightMapData[i])
    t_img = cv2.imread(terrainData[i])

    t_img = cv2.resize(t_img, (h_img.shape[1], h_img.shape[0]))
    combined = np.concatenate((h_img, t_img), axis = 1)
    cv2.imwrite(f"train/{i}.png", combined)
print("Done! combined heigh map with terrain")
```
