```python
from PIL import Image
import numpy as np
import os

base_path = r"C:\Users\yun\Desktop\CS591 digital image processing\final_project\Digital-Imaging-Final\TerrainHeightMap\Assets"

files = ["0002_h.png", "0003_h.png", "0007_h.png"]

for file in files:
    full_path = os.path.join(base_path, file)

    img = Image.open(full_path).convert("L")
    arr = np.array(img).astype(np.float32)

#    delete the extrem value
    p2, p98 = np.percentile(arr, (2, 98))
    arr = np.clip(arr, p2, p98)
    arr = (arr - p2) / (p98 - p2) # normalized

    arr = (arr * 255).astype(np.uint8)

    raw_name = os.path.join(base_path, os.path.splitext(file)[0] + ".raw")
    arr.tofile(raw_name)

    print(f"{file} → {raw_name} done")

print("Done! Transfor png into raw")
```
